```python
# import librairies and data of cereals

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/anisayari/Youtube-apprendre-python-intelligence-artificielle/master/%239%20-%20Data%20Exploration/cereal.csv')
df.head()
```

```python
df0 = df.copy()
```

```python
df0['type'].value_counts()
```

```python
import seaborn as sns
sns.set_theme(style="ticks")


sns.pairplot(df0[['calories', 'protein', 'fat', 'sodium', 'fiber',
       'carbo', 'sugars', 'potass', 'vitamins','type']], hue="type")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks")

plt.figure(figsize=(10,10))

sns.pairplot(df0[['calories', 'protein', 'fat', 'sodium', 'fiber',
       'carbo', 'sugars', 'potass', 'vitamins','type']], hue="type")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="dark")

plt.figure(figsize=(5,5))
sns.pairplot(df0[['calories', 'protein', 'fat', 'sugars','vitamins','shelf']], hue="shelf")
```

```python
# Distrubtion of gras

sns.countplot(x="fat", hue="type", data=df0)
```

```python
import plotly.express as px
fig = px.scatter(df0, x="sugars", y="rating", color='shelf')
fig.show()
```

```python
sns.regplot( x="sugars", y="rating", data=df0)
```

```python
sns.regplot( x="fat", y="rating", data=df0)
```

```python
sns.regplot( x="protein", y="rating", data=df0)
```

```python
df0["mfr"].value_counts().plot(kind="bar")
```

```python
sns.barplot(x="mfr", y="fat", hue="type", data = df0)
```

```python
sns.barplot(x="mfr", y="sugars", hue="type", data = df0)
```
