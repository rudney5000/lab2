### DATA OF POKEMON

```python
import pandas as pd
import seaborn as sns
sns.set(style ="whitegrid")
import plotly as plt
import matplotlib.pyplot as plt
```

```python
# import Data
df = pd.read_csv("https://raw.githubusercontent.com/anisayari/Youtube-apprendre-python-intelligence-artificielle/master/%238%20-%20Data%20Visualisation%20(Seaborn%2C%20Plotly)/pokemon.csv")
df
```

```python
# Info
df.info()
```

```python
df['capture_rate'].value_counts()
```

```python
df[df['capture_rate'] == '30 (Meteorite)255 (Core)']
```

```python
# np where()
df.loc[773, 'capture_rate'] = 30
df[df['name'] == 'Minior']
```

```python
df['capture_rate'] = df['capture_rate'].astype(int)
```

```python
df.info()
```

```python
# Describe
df.describe()
```

```python
# Data preparation

df.columns = df.columns.str.upper()
df['TYPE1'] = df['TYPE1'].str.capitalize()
df['TYPE2'] = df['TYPE2'].str.capitalize()
df['TYPE2'].fillna(value = 'None', inplace = True)
df.head()
```

```python
# Graph
sns.lineplot(x="DEFENSE", y= "ATTACK", data = df.sort_values(by = 'DEFENSE'))
```

```python
# Test Statistique
df.corr()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr())
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df[df['TYPE2']!='None'].groupby(['TYPE1', 'TYPE2']).size().unstack(), 
            linewidths=1, annot= True, cmap= sns.color_palette("RdBu", 7))
```

```python
#Plot

sns.relplot(
    x="DEFENSE", y="ATTACK", hue="GENERATION", size="WEIGHT_KG",
    sizes = (40, 400), alpha = .5, palette = "muted", height = 6, 
    data=df[df['GENERATION'].isin([1,7])]
)
```

```python
plt.figure(figsize=(20,8))
sns.boxplot(x="GENERATION", y ="WEIGHT_KG", data=df, palette="vlag", hue= "IS_LEGENDARY")
```

```python
#outsider pokemon

df[(df['GENERATION']==3) & (df['WEIGHT_KG']>800)]
```

```python

df[(df['GENERATION']==7) & (df['WEIGHT_KG']>900)]
```

```python
# Histogramme du nombre de pokemon par rapport au different generation
sns.countplot(x="GENERATION", data=df)
```

```python
# Diagramme en violon -catplot (kind= *violon*)
sns.set(style="whitegrid")


# Draw a nested barplot to show survival for class and sex

g = sns.catplot(x="GENERATION", y="HP", hue="IS_LEGENDARY", data=df,
                height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_ylabels("HP")
```

```python
# Displot
sns.displot(x="GENERATION", data=df, kind="kde")
```

```python

colours = ["cyan", "yellow"]
pokeLeg = df[df["IS_LEGENDARY"]==True]
pokeNon = df[df["IS_LEGENDARY"]==False]

legDist = [pokeLeg["NAME"].count(),pokeNon["NAME"].count()]
legPie = plt.pie(legDist,
                 labels = ["LEGENDARY", "NON LEGENDARY"],
                 autopct = "%1.1f%%",
                 shadow = True,
                 colors = colours,
                 startangle = 45,
                 explode=(0,0.1))
```

```python
import plotly.express as px
fig = px.line_polar(pd.DataFrame(dict(r=list(dict_tmp.values()),theta=list(dict_tmp.keys()))), r='r', theta='theta', line_close=True)
fig.update_layout(
    height=800,
    title_text=df.iloc[5].NAME
)
fig.show()
```

```python
# Nuage de point px.scater(df,x=NAME_column, y=NAME_column) fig.show()

fig = px.scatter(df, x="ATTACK", y="DEFENSE", 
                color="GENERATION", size="HP", 
                hover_data=['IS_LEGENDARY', 'NAME'])

fig.show()
```

```python
fig = px.scatter(df[df['GENERATION']==1], x="ATTACK", y="DEFENSE", 
                 color="IS_LEGENDARY", size="HP",
                 hover_data=['IS_LEGENDARY', 'NAME'])
fig.show()
```

```python
# line

import plotly.express as px
fig =px.line(df, x='WEIGHT_KG', y='BASE_HAPPINESS', color='GENERATION')
fig.show()
```

```python
# Boite a moustache - boxplot

fig = px.histogram(df, x='DEFENSE', y='ATTACK', color='IS_LEGENDARY',
                   marginal="box", hover_data=df.columns)
fig.show()
```

```python
import plotly.figure_factory as ff
import numpy as np

hist_data = [df.ATTACK.values, df.DEFENSE.values]

group_labels = ['ATTACK', 'DEFENSE']
colors = ['#393E46', '#2BCDC1', '#F66095']

fig = ff.create_distplot(hist_data, group_labels, colors=colors,
                         bin_size=[0.3, 0.2], show_curve=False)

# Add title
fig.update(layout_title_text='ATTACK/DEFENSE Histo')
fig.show()
```
